### Cotação Dolar Americano Comercial vs Real Brasileiro

```python
path_dolar = 'http://www.yahii.com.br/dolardiario21.html'

dolar_html = pd.read_html(path_dolar, encoding = 'ISO-8859-1', decimal = '.')
```

# Data Processing

### Combustivel Automotivo

```python
ca_columns = ['Regiao_Sigla', 'Estado_Sigla', 'Municipio', 'Revenda',
              'CNPJ_da_Revenda', 'Nome_da_Rua', 'Numero_Rua', 'Complemento', 'Bairro',
              'Cep', 'Produto', 'Data_da_Coleta', 'Valor_de_Venda', 'Valor_de_Compra',
              'Unidade_de_Medida', 'Bandeira']

ca_list = []
for i in ca_files:
    ca_unit = pd.read_csv(i, sep=';', encoding = 'ISO-8859-1', decimal = ',')
    ca_unit.rename(columns = dict(zip(ca_unit.columns,ca_columns)), inplace=True)
    ca_list.append(ca_unit)
    
df_ca = pd.concat(ca_list, ignore_index=True)
df_ca.rename(str.lower, axis='columns', inplace=True)
```

### Cotação Dolar Americano Comercial vs Real Brasileiro

```python
dolar_list = []

for i in np.arange(len(dolar_html)):
    if i > 3 and i < 16:
        dolar_unit = dolar_html[i]
        dolar_unit.columns = dolar_unit.iloc[1] 
        dolar_unit = dolar_unit[2:].reset_index(drop=True)
        dolar_list.append(dolar_unit)
        
    else:
        pass
    
df_dolar = pd.concat(dolar_list, ignore_index=True)
df_dolar.rename(str.lower, axis='columns', inplace=True)

df_dolar['compra'] = df_dolar['compra'].astype('float64')
df_dolar['venda'] = df_dolar['venda'].astype('float64')
df_dolar['data'] = df_dolar['data'].astype('datetime64')


df_dolar.loc[df_dolar['compra'] > 50, 'compra'] = df_dolar['compra'] / 10000
df_dolar.loc[df_dolar['venda'] > 50, 'venda'] = df_dolar['venda'] / 10000
```

### Preço Barril Petroleo Bruto tipo Brent (US$)

```python
df_p_brent.columns = df_p_brent.iloc[0]
df_p_brent = df_p_brent[1:].reset_index(drop=True)

df_p_brent['Preço - petróleo bruto - Brent (FOB)'] = df_p_brent['Preço - petróleo bruto - Brent (FOB)'].astype(int) / 100
df_p_brent['Data'] = df_p_brent['Data'].astype('datetime64')

p_brent_columns = ['data', 'preço_pretoleo_bruto_brent_fob US$']
df_p_brent.rename(columns = dict(zip(df_p_brent.columns,p_brent_columns)), inplace=True)

df_p_brent = df_p_brent.query('data >= "2021-01-01" & data <= "2021-12-31"')
```

```python
df_brent_real = df_p_brent.merge(df_dolar, left_on = 'data', right_on = 'data', sort=True)
df_brent_real['preço_pretoleo_bruto_brent_fob R$'] = df_brent_real['preço_pretoleo_bruto_brent_fob US$'] * df_brent_real['compra']
```
